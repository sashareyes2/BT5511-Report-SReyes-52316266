```python
import glob
import pandas as pd

# Get a list of filenames matching the pattern (e.g., all text files in the current directory)
expression_data = sorted(glob.glob("Genes/*.txt")) 

# Provide an empty list to store DataFrames
dfs = []

# Iterate over each file which allows access and manipulation of data in each row of the DataFrame sequentially
for file in expression_data:
    # Read the file into a DataFrame
    df = pd.read_csv(file, delimiter='\t')  # Reads data from a csv file into a DataFrame, using the specified delimiter
    # Append the DataFrame to the list
    dfs.append(df)

# Concatenate all DataFrames into a single DataFrame
merged_df = pd.concat(dfs, ignore_index=True)

merged_df
```

```python
# Detect Mutation Types between WT and Mut sequences in merged_df DataFrame
```

```python
def detect_mutation_type(wildtype_seq, mutant_seq):
    """Detect the type of mutation between wildtype and mutant sequences."""
    if len(wildtype_seq) != len(mutant_seq):  # checks if the lengths of 2 sequences are equal
        if len(wildtype_seq) > len(mutant_seq): # if length not the same = INSERTION OR DELETION
            return 'Deletion'
        else:
            return 'Insertion'
    for i in range(len(wildtype_seq)): # loop iterates that both sequences are compared from the first position to the last
        if wildtype_seq[i] != mutant_seq[i]: # checks if the nucleotide at position 'i' one sequence is the different to the other sequence at the same 'i' position 
            return 'Substitution'
    return 'None'

# Creates lists to store gene names and mutation type
gene_names = []
mutation_types = []

# Processes over each row in the merged DataFrame
for idx, row in merged_df.iterrows():
    mutation_type = detect_mutation_type(row['WildType.Sequence'], row['Mutant.Sequence'])
    gene_names.append(row['Gene'])
    mutation_types.append(mutation_type)

# Create DataFrame with gene names and mutation types
mutation_df = pd.DataFrame({'Gene': gene_names, 'Mutation Type': mutation_types})

mutation_df
```

```python
# Convert objects in merged_df DataFrame to floats for calculated analyses
```

```python
# Identify numerical columns
numerical_columns = merged_df.columns[3:]  # Shows that the first three columns are non-numeric

# Convert numerical columns to floats
merged_df[numerical_columns] = merged_df[numerical_columns].astype(float)

merged_df
```

```python
# Calculate mean values for wildtype mRNA expression
merged_df['Mean_mRNA_WT'] = merged_df[['mRNA.Expression.WT.Rep1', 'mRNA.Expression.WT.Rep2', 'mRNA.Expression.WT.Rep3']].mean(axis=1)

# Compile table for wildtype mRNA expression including replicates
mRNA_WT_table = merged_df[['Gene', 'mRNA.Expression.WT.Rep1', 'mRNA.Expression.WT.Rep2', 'mRNA.Expression.WT.Rep3', 'Mean_mRNA_WT']]

mRNA_WT_table
```

```python
# Calculate mean values for mutant mRNA expression
merged_df['Mean_mRNA_Mut'] = merged_df[['mRNA.Expression.Mut.Rep1', 'mRNA.Expression.Mut.Rep2', 'mRNA.Expression.Mut.Rep3']].mean(axis=1)

# Compile table for mutant mRNA expression including replicates
mRNA_Mut_table = merged_df[['Gene', 'mRNA.Expression.Mut.Rep1', 'mRNA.Expression.Mut.Rep2', 'mRNA.Expression.Mut.Rep3', 'Mean_mRNA_Mut']]

mRNA_Mut_table
```

```python
# Calculate mean values for wildtype protein expression
merged_df['Mean_Protein_WT'] = merged_df[['Protein.Expression.WT.Rep1', 'Protein.Expression.WT.Rep2', 'Protein.Expression.WT.Rep3']].mean(axis=1)

# Create separate tables for wildtype and mutant protein expression including replicates
protein_WT_table = merged_df[['Gene', 'Protein.Expression.WT.Rep1', 'Protein.Expression.WT.Rep2', 'Protein.Expression.WT.Rep3', 'Mean_Protein_WT']]

protein_WT_table
```

```python
# Calculate mean values for mutant protein expression
merged_df['Mean_Protein_Mut'] = merged_df[['Protein.Expression.Mut.Rep1', 'Protein.Expression.Mut.Rep2', 'Protein.Expression.Mut.Rep3']].mean(axis=1)

# Create separate tables for wildtype and mutant protein expression including replicates
protein_Mut_table = merged_df[['Gene', 'Protein.Expression.Mut.Rep1', 'Protein.Expression.Mut.Rep2', 'Protein.Expression.Mut.Rep3', 'Mean_Protein_Mut']]

protein_Mut_table
```

```python
# Calculate mean values for wildtype cell viability
merged_df['Mean_CellViability_WT'] = merged_df[['CellViability.WT.Rep1', 'CellViability.WT.Rep2', 'CellViability.WT.Rep3']].mean(axis=1)

# Create separate tables for wildtype and mutant cell viability including replicates
cellViability_WT_table = merged_df[['Gene', 'CellViability.WT.Rep1', 'CellViability.WT.Rep2', 'CellViability.WT.Rep3', 'Mean_CellViability_WT']]

cellViability_WT_table
```

```python
# Calculate mean values for mutant cell viability
merged_df['Mean_CellViability_Mut'] = merged_df[['CellViability.Mut.Rep1', 'CellViability.Mut.Rep2', 'CellViability.Mut.Rep3']].mean(axis=1)

# Compile table mutant cell viability including replicates
cellViability_Mut_table = merged_df[['Gene', 'CellViability.Mut.Rep1', 'CellViability.Mut.Rep2', 'CellViability.Mut.Rep3', 'Mean_CellViability_Mut']]

cellViability_Mut_table
```

```python
Mean_mRNA_WT = mRNA_WT_table['Mean_mRNA_WT']
Mean_mRNA_Mut = mRNA_Mut_table['Mean_mRNA_Mut']
Mean_Protein_WT = protein_WT_table['Mean_Protein_WT']
Mean_Protein_Mut = protein_Mut_table['Mean_Protein_Mut']
Mean_CellViability_WT = cellViability_WT_table['Mean_CellViability_WT']
Mean_CellViability_Mut = cellViability_Mut_table['Mean_CellViability_Mut']

# Append gene name, mutation type, and mean values for wildtype and mutant expression and viability levels
new_df = pd.DataFrame({
    'Gene': gene_names,
    'Mutation Type': mutation_types,
    'mRNA Expression WT Mean': Mean_mRNA_WT,
    'mRNA Expression Mut Mean': Mean_mRNA_Mut,
    'Protein Expression WT Mean': Mean_Protein_WT,
    'Protein Expression Mut Mean': Mean_Protein_Mut,
    'Cell Viability WT Mean': Mean_CellViability_WT,
    'Cell Viability Mut Mean': Mean_CellViability_Mut})

new_df
```

```python
# Calculate the difference in mean values between WT and Mut for mRNA expression
mRNA_difference = new_df['mRNA Expression WT Mean'] - new_df['mRNA Expression Mut Mean']

# Calculate the difference in mean values between WT and Mut for protein expression
protein_difference = new_df['Protein Expression WT Mean'] - new_df['Protein Expression Mut Mean']

# Calculate the difference in mean values between WT and Mut for cell viability
viability_difference = new_df['Cell Viability WT Mean'] - new_df['Cell Viability Mut Mean']

# Create a new DataFrame to store the differences
difference_df = pd.DataFrame({
    'Gene': new_df['Gene'],
    'Mutation Type': new_df['Mutation Type'],
    'mRNA Expression Difference': mRNA_difference,
    'Protein Expression Difference': protein_difference,
    'Cell Viability Difference': viability_difference})

difference_df
```

```python
# Creates a function to categorise the differences as increase, decrease, or no effect
def categorise_difference(difference):
    if difference > 0:
        return 'Increase'
    elif difference < 0:
        return 'Decrease'
    else:
        return 'No Effect'

# Apply the categorise_difference function to each difference column
mRNA_effect_on_WT = mRNA_difference.apply(categorise_difference)
protein_effect_on_WT = protein_difference.apply(categorise_difference)
viability_effect_on_WT = viability_difference.apply(categorise_difference)

# Create a new DataFrame to store the categorised differences
effect_df = pd.DataFrame({
    'Gene': difference_df['Gene'],
    'Mutation Type': difference_df['Mutation Type'],
    'mRNA Expression Difference': mRNA_difference,
    'mRNA Expression Effect': mRNA_effect_on_WT,
    'Protein Expression Difference': protein_difference,
    'Protein Expression Effect': protein_effect_on_WT,
    'Cell Viability Difference': viability_difference,
    'Cell Viability Effect': viability_effect_on_WT})

effect_df
```

```python
# DataFrame containing wildtype and mutant mRNA sequences in 'merged_df'

# To calculate the length of a sequence
def calculate_sequence_length(sequence):
    return len(sequence)

# Apply the function to calculate the length of each wildtype mRNA sequence
merged_df['mRNA Length WT'] = merged_df['WildType.Sequence'].apply(calculate_sequence_length)
merged_df['mRNA Length Mut'] = merged_df['Mutant.Sequence'].apply(calculate_sequence_length)

# Display the DataFrame with mRNA lengths
merged_df[['Gene', 'mRNA Length WT', 'mRNA Length Mut']]
```

```python
# To calculate the mutation position and location
def calculate_mutation_position_and_location(wildtype_sequence, mutant_sequence):
    for i in range(min(len(wildtype_sequence), len(mutant_sequence))):
        if wildtype_sequence[i] != mutant_sequence[i]:
            mutation_position = i + 1  # Mutation position (1-indexed)
            if mutation_position <= 1000:
                return mutation_position, 'Promoter'  # Mutation in promoter region
            else:
                return mutation_position, 'Coding sequence'  # Mutation in coding sequence
    return None, None  # If no mutation is found

# Apply the function to calculate the mutation position and location for each gene and 'expand' to multiple columns
merged_df[['Mutation Position', 'Mutation Location']] = merged_df.apply(lambda row: calculate_mutation_position_and_location(row['WildType.Sequence'], row['Mutant.Sequence']), axis=1, result_type='expand')


# Display the DataFrame with mRNA lengths and mutation information
merged_df[['Gene', 'mRNA Length WT', 'mRNA Length Mut', 'Mutation Position', 'Mutation Location']]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Create subplots using subplot_mosaic
fig, ax = plt.subplot_mosaic([["box", "scatter"], ["kde", "kde"]],
                             figsize=(14, 10), layout="constrained")

# Box plot
sns.boxplot(data=new_df, x='Mutation Type', y='mRNA Expression WT Mean', hue='Mutation Type', ax=ax['box'])
ax['box'].set_title('mRNA Expression WT Mean by Mutation Type')
ax['box'].set_xlabel('Mutation Type')
ax['box'].set_ylabel('mRNA Expression WT Mean')
ax['box'].set_yscale('log')
ax['box'].get_yaxis().set_major_formatter(plt.ScalarFormatter())

# Scatter plot
sns.scatterplot(data=new_df, x='mRNA Expression Mut Mean', y='Protein Expression Mut Mean', hue='Mutation Type', ax=ax['scatter'])
ax['scatter'].set_title('Protein Expression Mut Mean by mRNA Expression Mut Mean')
ax['scatter'].set_xlabel('mRNA Expression Mut Mean')
ax['scatter'].set_ylabel('Protein Expression Mut Mean')
ax['scatter'].set_xscale('log')
ax['scatter'].set_yscale('log')
ax['scatter'].get_xaxis().set_major_formatter(plt.ScalarFormatter())
ax['scatter'].get_yaxis().set_major_formatter(plt.ScalarFormatter())

# KDE plot for Cell Viability
sns.kdeplot(data=new_df, x='Cell Viability Mut Mean', hue='Mutation Type', ax=ax['kde'])
ax['kde'].set_title('Cell Viability Mut Mean Distribution by Mutation Type')
ax['kde'].set_xlabel('Cell Viability Mut Mean')
ax['kde'].set_ylabel('Density')

plt.show()
```

```python
# Calculate the absolute difference in Cell Viability between wildtype and mutant
new_df['Cell Viability Difference'] = abs(new_df['Cell Viability WT Mean'] - new_df['Cell Viability Mut Mean'])

# Sort the DataFrame by cell viability difference in descending order
sorted_df = new_df.sort_values(by='Cell Viability Difference', ascending=False)

# Get the top 5 genes and mutations
top_5_genes_mutations_cell_viability = sorted_df[['Gene', 'Mutation Type', 'Cell Viability Difference']].head(5)

# Display the top 5 genes and mutations
top_5_genes_mutations_cell_viability
```

```python
# Calculate the absolute difference in mRNA Expression between wildtype and mutant
new_df['mRNA Expression Difference'] = abs(new_df['mRNA Expression WT Mean'] - new_df['mRNA Expression Mut Mean'])

# Sort the DataFrame by mRNA Expression difference in descending order
sorted_df = new_df.sort_values(by='mRNA Expression Difference', ascending=False)

# Get the top 5 genes and mutations
top_5_genes_mutations_mRNA_difference = sorted_df[['Gene', 'Mutation Type', 'mRNA Expression Difference']].head(5)

# Display the top 5 genes and mutations
top_5_genes_mutations_mRNA_difference

```

```python
# Calculate the absolute difference in Protein Expression between wildtype and mutant
new_df['Protein Expression Difference'] = abs(new_df['Protein Expression WT Mean'] - new_df['Protein Expression Mut Mean'])

# Sort the DataFrame by Protein Expression difference in descending order
sorted_df = new_df.sort_values(by='Protein Expression Difference', ascending=False)

# Get the top 5 genes and mutations
top_5_genes_mutations_protein_difference = sorted_df[['Gene', 'Mutation Type', 'Protein Expression Difference']].head(5)

# Display the top 5 genes and mutations
top_5_genes_mutations_protein_difference
```

```python
# Extract top 5 genes from each dataset
top_5_cell_viability = set(top_5_genes_mutations_cell_viability['Gene'])
top_5_mRNA_difference = set(top_5_genes_mutations_mRNA_difference['Gene'])
top_5_protein_difference = set(top_5_genes_mutations_protein_difference['Gene'])

# Combine top 5 genes from all three datasets
combined_genes = top_5_cell_viability.union(top_5_mRNA_difference, top_5_protein_difference)

# Count occurrences of each gene in each dataset
cell_viability_counts = {gene: 0 for gene in combined_genes}
mRNA_difference_counts = {gene: 0 for gene in combined_genes}
protein_difference_counts = {gene: 0 for gene in combined_genes}

# Iterate through the datasets and count occurrences of each gene
for gene in combined_genes:
    if gene in top_5_cell_viability:
        cell_viability_counts[gene] += 1
    if gene in top_5_mRNA_difference:
        mRNA_difference_counts[gene] += 1
    if gene in top_5_protein_difference:
        protein_difference_counts[gene] += 1

# Plotting
plt.figure(figsize=(10, 6))

# Plot bars for each dataset
plt.barh(list(combined_genes), [cell_viability_counts[gene] for gene in combined_genes], color='skyblue', label='Cell Viability')
plt.barh(list(combined_genes), [mRNA_difference_counts[gene] for gene in combined_genes], color='salmon', left=[cell_viability_counts[gene] for gene in combined_genes], label='mRNA Difference')
plt.barh(list(combined_genes), [protein_difference_counts[gene] for gene in combined_genes], color='lightgreen', left=[cell_viability_counts[gene] + mRNA_difference_counts[gene] for gene in combined_genes], label='Protein Difference')

# Add legend and labels
plt.xlabel('Frequency')
plt.ylabel('Genes')
plt.title('Top 5 Genes in Top Mutations from all three datasets')
plt.legend()

plt.tight_layout()
plt.show()
```
